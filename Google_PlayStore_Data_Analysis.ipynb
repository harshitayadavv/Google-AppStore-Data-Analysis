```python
# Importing necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('googleplaystore.csv')
df
```

## Data Cleaning

```python
# Renaming the 'Unnamed: 0' column to 'index': This renames the column for clarity, making it more meaningful for analysis.
df = df.rename(columns={'Unnamed: 0': 'index'})
df
```

```python
# Are there any missing or null values in the dataset?
# This will check if any values are missing or null across the dataset, helping to ensure completeness.
missing_values = df.isnull().sum()
print("Missing values in each column:\n", missing_values)
```

```python
# Are there any duplicates in the 'app' column?
duplicate_apps = df[df.duplicated(subset='App', keep=False)]
duplicate_apps_info = duplicate_apps[['App']].reset_index()
print("Duplicate Apps (Index and App Name):\n", duplicate_apps_info)
```

```python
# Check if 'Rating' lies between 0 and 5, and replace invalid values with 0
df['Rating'] = pd.to_numeric(df['Rating'], errors='coerce')  # Convert to numeric, forcing errors to NaN
df['Rating'] = df['Rating'].apply(lambda x: x if 0 <= x <= 5 else 0)  # Replace values outside 0 to 5 with 0

```

```python
# Do the 'reviews' contain valid numeric values?
df['Reviews'] = df['Reviews'].fillna(0)  # Replace NaN with 0
```

```python
# Are there any outliers or incorrect values in the 'category' and 'genre' columns?
df['Category'] = df['Category'].str.strip()  # Remove any leading/trailing spaces
df['Genres'] = df['Genres'].str.strip()  # Remove any leading/trailing spaces
```

## Data Analysis

```python
# 1. What is the average rating per category?
avg_rating_per_category = df.groupby('Category')['Rating'].mean()
print("Average rating per category:\n", avg_rating_per_category)
```

```python
# 2. What is the relationship between app rating and the number of reviews?

# Function to clean the 'Reviews' column (remove 'M' and 'K', convert to numeric values)
def clean_reviews(review_str):
    if isinstance(review_str, str):
        if 'M' in review_str:
            return float(review_str.replace('M', '').replace(',', '')) * 1e6  # Convert millions to actual number
        elif 'K' in review_str:
            return float(review_str.replace('K', '').replace(',', '')) * 1e3  # Convert thousands to actual number
        else:
            return float(review_str.replace(',', ''))  # For other cases
    return review_str

# Apply the clean_reviews function to the 'Reviews' column
df['Reviews'] = df['Reviews'].apply(clean_reviews)

# Convert the 'Rating' column to numeric (if it's not already)
df['Rating'] = pd.to_numeric(df['Rating'], errors='coerce')

# Creating Review Categories based on the number of reviews
bins = [0, 1e3, 1e4, 1e5, 1e6, 1e7, 1e8, np.inf]
labels = ['<1K', '1K-10K', '10K-100K', '100K-1M', '1M-10M', '10M-100M', '>100M']
df['Review Category'] = pd.cut(df['Reviews'], bins=bins, labels=labels)

# Calculate the average rating per review category with observed=True to avoid the warning
average_rating_by_reviews = df.groupby('Review Category', observed=True)['Rating'].mean()

# Print the average rating per review category
print("\nAverage Rating per Review Category:\n", average_rating_by_reviews)
```

```python
# 3. Which category has the highest average number of installs?

# Clean the 'Installs' column by removing non-numeric characters like ',' and '+'
df['Installs'] = df['Installs'].replace({'[,+]': ''}, regex=True)  # Remove commas and plus sign
df['Installs'] = pd.to_numeric(df['Installs'], errors='coerce')  # Convert to numeric, force errors to NaN

# Now calculate the average installs per category
avg_installs_per_category = df.groupby('Category')['Installs'].mean()

# Find the category with the highest average installs
category_with_highest_installs = avg_installs_per_category.idxmax()
print("Category with highest average installs:", category_with_highest_installs)
```

```python
# 4. What is the most common content rating?
most_common_content_rating = df['Content Rating'].mode()[0]
print("Most common content rating:", most_common_content_rating)
```

```python
# 5. How does the app price correlate with the number of installs?

# Clean the 'Price' column by removing the dollar sign and spaces, and convert it to numeric
df['Price'] = df['Price'].replace({r'\$': '', ' ': ''}, regex=True)  # Remove dollar sign and spaces
df['Price'] = pd.to_numeric(df['Price'], errors='coerce')  # Convert to numeric, force errors to NaN

# Now calculate the correlation between Price and Installs
price_installs_correlation = df[['Price', 'Installs']].corr()
print("Correlation between Price and Installs:\n", price_installs_correlation)

```

```python
# 6. Which category has the highest number of paid apps?
paid_apps_by_category = df[df['Type'] == 'Paid'].groupby('Category').size()
category_with_highest_paid_apps = paid_apps_by_category.idxmax()
print("Category with the highest number of paid apps:", category_with_highest_paid_apps)
```

```python
# 7. What is the most common Android version required by apps?
most_common_android_version = df['Android Ver'].mode()[0]
print("Most common Android version required by apps:", most_common_android_version)
```

```python
# 8. What are the top 10 most popular apps by installs?
top_10_apps_by_installs = df.sort_values(by='Installs', ascending=False).head(10)
print("Top 10 most popular apps by installs:\n", top_10_apps_by_installs[['App', 'Installs']])

```

```python
# 10. How many apps are free, and how many are paid?
free_apps_count = df[df['Type'] == 'Free'].shape[0]
paid_apps_count = df[df['Type'] == 'Paid'].shape[0]
print(f"Number of free apps: {free_apps_count}")
print(f"Number of paid apps: {paid_apps_count}")
```

### Data Visualisation

```python
### 1. What is the distribution of app ratings across all apps?
plt.figure(figsize=(8, 6))
sns.histplot(df['Rating'], bins=20, kde=True, color='blue')
plt.title('Distribution of App Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```

```python
### 2. What is the distribution of app sizes?
plt.figure(figsize=(10, 6))
sns.histplot(df['Size'], bins=20, kde=True, color='green')
plt.title('Distribution of App Sizes')
plt.xlabel('Size')
plt.ylabel('Frequency')
plt.show()
```

```python
### 3. How are the number of installs distributed among apps?
plt.figure(figsize=(10, 6))
sns.histplot(df['Installs'], bins=20, kde=True, color='red')
plt.title('Distribution of Number of Installs')
plt.xlabel('Number of Installs')
plt.ylabel('Frequency')
plt.show()
```

```python
### 4. What is the relationship between app rating and number of reviews?
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Rating', y='Reviews', color='purple')
plt.title('Relationship Between Rating and Reviews')
plt.xlabel('Rating')
plt.ylabel('Number of Reviews')
plt.show()
```

```python
### 5. How does the number of installs vary by category?
category_installs = df.groupby('Category')['Installs'].mean().sort_values(ascending=False)
plt.figure(figsize=(10, 6))
category_installs.plot(kind='bar', color='orange')
plt.title('Average Number of Installs by Category')
plt.xlabel('Category')
plt.ylabel('Average Number of Installs')
plt.xticks(rotation=90)
plt.show()
```

```python
### 6. What is the distribution of app prices?
plt.figure(figsize=(10, 6))
sns.histplot(df['Price'], bins=20, kde=True, color='pink')
plt.title('Distribution of App Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
### 7. What percentage of apps are free versus paid?
type_counts = df['Type'].value_counts()
plt.figure(figsize=(8, 6))
type_counts.plot(kind='pie', autopct='%1.1f%%', colors=['lightblue', 'lightcoral'])
plt.title('Percentage of Free vs Paid Apps')
plt.ylabel('')
plt.show()
```

```python
### 8. How does the average rating compare between free and paid apps?
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Type', y='Rating', hue='Type', palette='Set2', legend=False)
plt.title('Rating Distribution Between Free and Paid Apps')
plt.xlabel('App Type')
plt.ylabel('Rating')
plt.show()
```

```python
### 9. Which category has the highest average rating?
category_ratings = df.groupby('Category')['Rating'].mean().sort_values(ascending=False)
plt.figure(figsize=(10, 6))
category_ratings.plot(kind='bar', color='brown')
plt.title('Average Rating by Category')
plt.xlabel('Category')
plt.ylabel('Average Rating')
plt.xticks(rotation=90)
plt.show()
```

```python
### 10. What is the relationship between app price and rating?
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Price', y='Rating', color='cyan')
plt.title('Price vs Rating of Apps')
plt.xlabel('Price')
plt.ylabel('Rating')
plt.show()
```

```python
# 13. What is the distribution of content ratings?
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Content Rating', hue='Content Rating', palette='Set3', legend=False)
plt.title('Distribution of Content Ratings')
plt.xlabel('Content Rating')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()
```

```python
### 14. Which category has the highest number of installs in total?
category_total_installs = df.groupby('Category')['Installs'].sum().sort_values(ascending=False)
plt.figure(figsize=(10, 6))
category_total_installs.plot(kind='bar', color='lightgreen')
plt.title('Total Installs by Category')
plt.xlabel('Category')
plt.ylabel('Total Installs')
plt.xticks(rotation=90)
plt.show()
```

```python
### 15. How do the top-rated apps compare to the most installed apps?
top_rated_apps = df.nlargest(20, 'Rating')
most_installed_apps = df.nlargest(20, 'Installs')
plt.figure(figsize=(10, 6))
sns.scatterplot(x=top_rated_apps['Installs'], y=top_rated_apps['Rating'], color='blue', label='Top Rated Apps')
sns.scatterplot(x=most_installed_apps['Installs'], y=most_installed_apps['Rating'], color='red', label='Most Installed Apps')
plt.title('Top Rated Apps vs Most Installed Apps')
plt.xlabel('Number of Installs')
plt.ylabel('Rating')
plt.legend()
plt.show()

```
